```python
# Titanic Survival Prediction
### Data Science Project
```

```python
## 1. Import Libraries
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## 2. Load Dataset
```

```python
data = pd.read_csv("titanic.csv")
```

```python
data.head()
data.isnull().sum()
```

```python
## 3. Data Cleaning
```

```python
# Handle missing values
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Fare'] = data['Fare'].fillna(data['Fare'].mean())

# Convert categorical to numeric
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
```

```python
## 4. Data Visualization
```

```python
sns.countplot(x='Survived', data=data)
plt.show()
```

```python
## 5. Feature Selection
```

```python
x= data[['Pclass', 'Sex', 'Age', 'Fare']]
y =data['Survived']
```

```python
## 6. Train-Test Split
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
X_train = X_train.fillna(X_train.mean())
X_test = X_test.fillna(X_test.mean())
```

```python
## 7. Model Training
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
## 8. Model Evaluation
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
## 9. Conclusion

In this project, a Logistic Regression model was trained on the Titanic dataset
to predict passenger survival. After data cleaning, visualization, and feature
selection, the model achieved an accuracy of approximately 80%.
```
